## Poisson Distribution

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import math
```

#### Create a function to compute the poisson distribution

```python
# k -> no of occurence of an event.
# lam(lambda) - expected no of occurence of an event in a time frame.

def poisson_dist(k, lam):
    # formula below to compute poisson distribution.
    dist = math.pow(lam, k)*np.exp(-lam)/math.factorial(k)
    return dist
```

#### Poisson Distribution Formula

#### Next we need to find out the probabaility of getting 10 complaints in a day given that expected complaints in a regular day is 8.

k = 10, (probability of 10 complaint in a day),

lam = 8 (8 complaints expected in a regular day)

```python
poisson_dist(10, 8)
```

We got **0.0992** as distribution output. This means that we gonna get **9.92%** of chance of getting
10 complaints per day.
This is what poisson distribution gives. 

```python
0.0992*100
```

### Alternate Way to find Poisson Distribution

```python
from scipy.stats import poisson
```

```python
# pass k and lambda value to pmf(probabaility mass function) as args.
dist = poisson.pmf(10, 8)
dist
```

```python
dist*100
```

We get same distribution as previous method.

9.92% of chance for getting 10 complaints in a regular day.

### Plot the Poisson Distribution

Plot the poisson distribution to find out the most probable number of complaints.

```python
lam = 6 # lambda, expected # of complaints in a day
k_vals = np.arange(0, 20) # probability of # complaints in a day
dist = np.zeros(len(k_vals)) # probability distribution of zeros with size of k_vals array.

# test distribution for different values of k.
for k in range(len(k_vals)):
    dist[k]=poisson.pmf(k, lam)
```

```python
# probability distributons
dist
```

```python
# Plot the above distribution
# pass k values and distribution to bar plot as x and y axis.

plt.bar(k_vals, dist)
plt.xlabel("Complaints")
plt.ylabel("Days")
plt.title(label="Poisson Distribtion")
plt.show()
```

We can infer from the plot that, 5 - 6 complaints were registered as the most probable complaints in a regular day.
